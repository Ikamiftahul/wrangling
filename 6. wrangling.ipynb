## Data Wrangling and Manipulation with Python

### What is Data Wrangling?

Data wrangling is the process of collecting, and modifying the raw data into anotherform for analyzing and decision-making easily. It is also known as Data Munging. For example, when we wanr only some part of the data that is useful based on the aopplication, then we can do data wrangling. This process is widlt used in the data science domain.

### Why is Data Wrangling Important?

1. Data Quality: Raw data often contains errors, missing values, and inconsistencies. Data wrangling improves data quality
2. Analysis Readiness: Cleaned and structured data is essential for data analysis and machine learning.
3. Time Efficiency: Properly wrangled data reduces analysis time and leads to faster insights.

### Data Wrangling Workflow Overview

Data wrangling typically follows these steps:

1. Data Collection: Obtain the raw data from various sourcesm such as databases, spreadsheets, or APIs
2. Data Cleaning: Clean the data by addressing missing values: duplicates, and errors.
3. Data Transformation: Transform the data by filtering, sorting, aggregating, and reshaping it
4. Data Encrichment: Create derived features or enrich the data with external sources.
5. Data Exploration: Explore and visualize the data to gain insights.
6. Data Storage: Store the cleaned and transformed data for analysis or modeling

Using the `pandas` and `numpy` modules, we can do the below operation for data wrangling:

1. Handling missing or null values
2. Grouping Data
3. Reshaping the data: In this process, data is manipulated according to the requirements, where new data can be added or pre-existing data can be modified
4. Filtering the data: SOmetimes datasets are composed of unwanted rows or columns which are required to be removed or filtered, etc

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# load the Titanic dataset as a CSV file
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
# check the first few rows of the dataset
df.head()
```

```python
# check data types / info
df.info()
```

```python
# check missing values
df.isna().sum()
```

```python
# get basic statistics of numeric columns
df.describe()
```

# Data Transformation
**Data transformation involes various operations to reshape and modify data to suit your analytical needs**

```python
# Filtering and subsetting data, allows you to select spesific rows or columns based on conditions

# select passengers who survived
survived_passengers = df[df['Survived'] == 1]

# select columns of interest
selected_columns = df[['Name', 'Age', 'Sex']]
```

```python
# survived passengers
survived_passengers.head()
```

```python
# selected columns
selected_columns.head()
```

```python
# Sorting, helps in arranging data in ascending or descending order

# sort by age in descending order
sorted_by_age = df.sort_values(by='Age', ascending=False)
```

```python
# sorted data
sorted_by_age.head()
```

```python
# Aggregation, combines multiple data points into summary statistic

# calculating the average age of passengers
df['Age'].mean()
```

```python
# calculate the max age of passengers
df["Age"].agg("max")
```

```python
# Reshaping data involves changing its structure, e.g., from wide to long format

# Reshape data from wide to long format
melted_data = pd.melt(df, id_vars=['PassengerId', 'Survived'], var_name='Variable', value_name='Value') # using pd.melt()
```

```python
# reshaped data
melted_data.head()
```

```python
# Creating new features from existing ones

# create a new feature 'FamilySize' by adding 'SibSp' and 'Parch'
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
# new feature created
df.head()
```

## Data Grouping and Summarization
Data grouping helps in summarizing data by one or more categorical variables

```python
# Group data by 'Pclass' and calculate the mean age for each class
class_age_menas = df.groupby('Pclass')['Age'].mean()
```

```python
# Grouped data
class_age_menas
```

```python
# Aggregate functions (e.g., sum, mean, count) to grouped data

# calculate the total fare for each class
class_fare_totals = df.groupby('Pclass')['Fare'].sum()
```

```python
# Grouped data
class_fare_totals
```

```python
# Multi-level Grouping

# Grou data by 'Pclass', 'Sex', and 'Embarked' and calculate the mean age for  each group
class_sex_embarked_age_means = df.groupby(['Pclass', 'Sex', 'Embarked'], as_index=False)['Age'].mean()
```

```python
# Multi grouped data
class_sex_embarked_age_means
```

```python
# Plot tables allow for more advenced data summarization

# pivat table to see the survival rate by class and sex
pivot_table = pd.pivot_table(df, values='Age', index='Pclass', columns='Sex', aggfunc='mean')

# create a pivot table to see the survival rate by class, sex, and age
pivot_table2 = pd.pivot_table(df, values='Age', index=['Pclass', 'Sex'], columns=pd.cut(df['Age'], [0, 18, 30, 50, 100]), aggfunc='mean')
```

```python
# pivot 1
pivot_table
```

```python
# pivot 2
pivot_table2
```

## Data Merging and Joining

**Data merging involes combining multiple datasets into one**

```python
# split data based on Pclass
df1 = df[df["Pclass"] == 1]
df2 = df[df["Pclass"] == 2]
```

```python
df1.shape, df2.shape
```

```python
# combining DataFrames using pd.concat()

# Concatenate vertically
combined_data = pd.concat([df1, df2])

# Concatenate horizontally
combined_data2 = pd.concat([df1, df2], axis=1)
```

```python
# Vertically concatenated
combined_data.reset_index(drop=True).head()
```

```python
# Horizontally concatenated
combined_data2.reset_index(drop=True).head()
```

```python
# Merge two DataFrame based on a common column ('PassengerId') using an inner join
merged_data = pd.merge(df1, df2,
                      on = 'PassengerId', how='left',
                      suffixes=('_A','_B'))  # you can chenge the 'how' part to 'right', 'left', 'inner', 'outer'
```

```python
# merged data
merged_data.head()
```

## Data Storing

**You can store the data after doing data wrangling using [pandas library](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html)**

```python
# prepare the data
combined_data = combined_data.reset_index(drop=True)
```

```python
# export the prepared data to csv
combined_data.to_csv("result/titanic_cleaned.csv")
```

```python
# export the prepared data to json
combined_data.to_json("result/titanic_cleaned.json")
```
